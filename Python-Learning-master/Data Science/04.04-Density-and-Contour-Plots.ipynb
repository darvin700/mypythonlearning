# Density and Contour Plots


## What Are Density and Contour Plots?

### **Density Plot**:
- A density plot shows the **concentration of data points** in an area.
- Think of it as showing "**how crowded**" the data is in different regions.

### **Contour Plot**:
- A contour plot is like a **map for elevation**, where lines connect points of the same value.
- Each line (contour) shows where the value is constant.

---

## Everyday Analogy
- Imagine a **hilly terrain**:
  - A **density plot** would shade areas darker where the hills are taller (showing more height or "crowded" height).
  - A **contour plot** would draw lines around each hill at the same height, like on a topographic map.

---

## Why Use These Plots?

### **Density Plot**:
- Helps to see the **"busy" or "empty" regions** in your data.

### **Contour Plot**:
- Useful for visualizing how values **change smoothly over a surface**.

---

## How to Create Density and Contour Plots?
We will use Python's **Matplotlib** library for examples.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
x = np.random.normal(size=1000)  # 1000 random x-coordinates
y = np.random.normal(size=1000)  # 1000 random y-coordinates

# Create a density plot
plt.hist2d(x, y, bins=30, cmap='Blues')  # 2D histogram
# bins=30:
# Divides the x-axis and y-axis into 30 equally spaced intervals.
# The grid will have 30 × 30=90030×30=900 bins.

# cmap='Blues':
# Specifies the color map used to represent the density.

plt.colorbar(label='Density')  # Add a color bar to show density scale
plt.title("Density Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## Visualizing a Three-Dimensional Function

We'll start by demonstrating a contour plot using a function $z = f(x, y)$, using the following particular choice for $f$ (we've seen this before in [Computation on Arrays: Broadcasting](02.05-Computation-on-arrays-broadcasting.ipynb), when we used it as a motivating example for array broadcasting):

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

A contour plot can be created with the ``plt.contour`` function.
It takes three arguments: a grid of *x* values, a grid of *y* values, and a grid of *z* values.
The *x* and *y* values represent positions on the plot, and the *z* values will be represented by the contour levels.
Perhaps the most straightforward way to prepare such data is to use the ``np.meshgrid`` function, which builds two-dimensional grids from one-dimensional arrays:

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a grid of x and y values
# x and y are 100 equally spaced values between -3 and 3
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

# Print the x and y arrays to understand the grid values
print("X values (1D array):", x[:5], "...")  # Print first 5 x values
print("Y values (1D array):", y[:5], "...")  # Print first 5 y values

# Step 2: Create a 2D grid from x and y values
# Meshgrid creates 2D arrays for all combinations of x and y
X, Y = np.meshgrid(x, y)

# Print the first few rows of the grid to understand how it looks
# Print top-left 5x5 of X grid
print("\nX grid (2D array):\n", X[:5, :5], "...\n")
# Print top-left 5x5 of Y grid
print("Y grid (2D array):\n", Y[:5, :5], "...\n")

# Step 3: Define a function based on the grid values
# This function creates a "hill" shape using sine of the distance from the origin
Z = np.sin(np.sqrt(X**2 + Y**2))

# Print the first few values of Z to understand the height values
print("Z values (Height):\n", Z[:5, :5], "...")  # Print top-left 5x5 of Z grid

# Step 4: Create a contour plot
# Contour lines represent regions with the same height in the Z grid
plt.contour(X, Y, Z, levels=10, cmap='viridis')  # Contour lines with 10 levels

# Step 5: Add a color bar to visualize the height values
plt.colorbar(label='Height')  # The color bar shows how colors map to Z values

# Step 6: Add labels and title for clarity
plt.title("Contour Plot")  # Title of the plot
plt.xlabel("X-axis")  # Label for the horizontal axis
plt.ylabel("Y-axis")  # Label for the vertical axis

# Step 7: Show the plot
plt.show()
```

```python
# X-axis and Y-axis:
# These represent the grid of points (positions).
# Contour Circles:
# Each circle shows where the height is the same.
# Circles further out represent larger distances from the center.
# Colors:
# Help us understand the height (Z value) at each point.
```
